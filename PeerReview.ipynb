```python
import pandas as pd
from jinja2 import Template

# read data from CSV file into a pandas dataframe
df = pd.read_excel('CSV_W23.xlsx')

# group data by GroupNo
grouped = df.groupby('GroupNo')

# iterate through each group
for group, data in grouped:
    # calculate mean of integer columns for the group
    means = data[['Smooth_Presentation', 'Clear_Presentation', 'Explain_CBD', 'Balanced', 'Kin_Explaination', 'Doseage', 'SideEffects', 'Bias', 'Pros_Cons', 'Good_Job', 'Research_Holes', 'Peer_Reviewed', 'Sufficent_Info']].agg(['mean', 'std'])
    #means = data[['SmoothPresent', 'ClearPresent', 'Comment_ton', 'Explain_CBD', 'Balanced', 'Kin_Explain', 'Doseage', 'SideEffect', 'Bias', 'ProCon']].mean()
    
    # format mean and standard deviation values to two decimal places
    means = means.round(2)

    # extract only the mean values
    score = means.loc['mean']
    
    # Calculate total score
    total_score = score.sum()
    
    # Convert total score to a mark out of 120 as a percentage
    mark = round(total_score / 120 * 100, 2)
   
    print(group)
   
    # create individual report for the group
    template = Template("""
       <html>
    <head>
        <title>Group {{ group }}</title>
        <style>
            table {
              border-collapse: collapse;
              width: 80%;
              margin: 0 auto;
            }

            th, td {
              text-align: left;
              padding: 8px;
              border-bottom: 1px solid #ddd;
            }

            th {
              background-color: #f2f2f2;
            }

            body {
              max-width: 800px;
              margin: 0 auto;
              padding: 20px;
              font-family: sans-serif;
            }
        </style>
    </head>
    <body>
        <h1>Group {{ group }}</h1>
        <h2>Summary of Scores:</h2>
        <table>
            <thead>
                <tr>
                    <th>Section</th>
                    <th>Mean Score</th>
                    <th>Standard Deviation</th>
                </tr>
            </thead>
            <tbody>
                {% for column in means.columns %}
                <tr>
                    <td>{{ column }}</td>
                    <td>{{ means.loc['mean', column] }}</td>
                    <td>{{ means.loc['std', column] }}</td>
                </tr>
                {% endfor %}
            </tbody>
        </table>
        <h2>Peer Review Grade: {{ mark }} </h2>
        <h4>Note* Pros_Cons and Good_Job are scored out of 5 everything else is out of 10</h4>
        <h2>Feedback Presentation tone:</h2>
        {% for feedback in data['Comment_ton'] %}
        <p>{{ feedback }}</p>
        {% endfor %}
        <h2>Feedback Overal Presentation:</h2>
        {% for feedback in data['Comment_overal'] %}
        <p>{{ feedback }}</p>
        {% endfor %}
         <h2>Feedback Presentation Qaulity:</h2>
        {% for feedback in data['Comment_quality'] %}
        <p>{{ feedback }}</p>
        {% endfor %}
    </body>
</html>

    """)
    html = template.render(group=group, means=means, mark=mark, data=data)
    # save individual report as HTML file
    with open(f'group_{group}_report.html', 'w') as f:
        f.write(html)
   
```
